## 3. Basic Statistics and Distributions

```python
# Accounts by plan
accounts['plan_tier'].value_counts(normalize=True).plot(kind='bar', title='Accounts by Plan Tier')
```

```python
# Referral sources
accounts['referral_source'].value_counts().plot(kind='barh', title='Referral Source Distribution')
```

```python
# Churn rate
churn_rate = accounts['churn_flag'].mean()
print(f"Overall churn rate: {churn_rate:.2%}")
```

```python
# Industries
accounts['industry'].value_counts().head(10).plot(kind='bar', title='Top 10 Industries')
```

## 4. Time Trends

```python
# Signups by Month
accounts['signup_date'] = pd.to_datetime(accounts['signup_date'])
accounts.set_index('signup_date').resample('M').size().plot(title="Monthly Signups")
```

```python
# New subscriptions per month
subscriptions['start_date'] = pd.to_datetime(subscriptions['start_date'])
subscriptions.set_index('start_date').resample('M').size().plot(title="Subscription Starts Over Time")
```
